```python
from qiskit import QuantumRegister, ClassicalRegister, QuantumCircuit, execute, Aer
from random import randrange

qc_a = QuantumCircuit(2,2)

qc_a.x(0)
qc_a.x(1)

qc_a.cx(0,1)

qc_a.x(0)
qc_a.x(1)

qc_a.measure(0,1)
qc_a.measure(1,0)

qc_a.draw(output='mpl')
```

```python
job = execute(qc_a, Aer.get_backend('qasm_simulator'), shots = 1000)
counts = job.result().get_counts()
print(counts) # counts is a dictionary
```

```python
qc_b = QuantumCircuit(2,2)

qc_b.x(0)

qc_b.swap(0,1)

qc_b.x(0)
qc_b.x(1)

qc_b.measure(0,1)
qc_b.measure(1,0)

qc_b.draw(output='mpl')
```

```python
qc_c = QuantumCircuit(2,6)

qc_c.x(0)

qc_c.barrier()

qc_c.measure(0,0)

qc_c.cx(1,0)

qc_c.x(0)

qc_c.measure(1,1)
qc_c.measure(0,2)
qc_c.measure(1,4)
qc_c.measure(0,3)

qc_c.x(1)

qc_c.measure(1,5)

qc_c.draw(output='mpl')
```

```python
job = execute(qc_c, Aer.get_backend('qasm_simulator'), shots = 1000)
counts = job.result().get_counts()
print(counts) # counts is a dictionary
```
