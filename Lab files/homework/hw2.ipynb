```python
from qiskit import QuantumRegister, ClassicalRegister, QuantumCircuit, execute, Aer
from random import randrange

qc_a = QuantumCircuit(2,2)

qc_a.h(0)
qc_a.x(1)

qc_a.cx(0,1)

qc_a.swap(0,1)

qc_a.measure(0,1)
qc_a.measure(1,0)

qc_a.draw(output='mpl')
```

```python
qc_b = QuantumCircuit(2,4)

qc_b.i(0)

qc_b.measure(0,3)

qc_b.cx(1,0)

qc_b.measure(1,1)
qc_b.measure(0,2)

qc_b.x(1)

qc_b.measure(0,0)

qc_b.draw(output='mpl')
```

Qiskit Calculations

```python
qc_c = QuantumCircuit(2)

qc_c.x(1)

qc_c.cx(1,0)

qc_c.h(0)

qc_c.measure_all()

qc_c.draw(output='mpl')
```

```python
job = execute(qc_c, Aer.get_backend('qasm_simulator'), shots = 1000)
counts = job.result().get_counts()
print(counts) # counts is a dictionary
```

```python
qc_d = QuantumCircuit(2)

qc_d.h(0)
qc_d.h(1)

qc_d.cx(0,1)

qc_d.h(0)
qc_d.h(1)

qc_d.measure_all()

qc_d.draw(output='mpl')
```
